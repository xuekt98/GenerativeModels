```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_points(xs, ys, xlabel="x", ylabel="y", title="title"):
    plt.figure(figsize=(8,8))
    plt.plot(xs, ys, marker='o', c='purple', markersize=0.1)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
beta_min = 0.0001
beta_max = 0.0095
T = 200
steps = np.arange(T)
betas = np.linspace(beta_min, beta_max, 200)
show_points(steps, betas, "steps", "betas", "betas")
```

```python
alphas = 1. - betas
show_points(steps, alphas, "steps", "alphas", "alphas")
```

```python
beta_min = 0.0001
beta_max = 0.00267
T = 500
steps = np.arange(T)
betas = np.linspace(beta_min, beta_max, 500)
alphas = 1. - betas
alphas_cumprod = np.cumprod(alphas, axis=0)
# print(alphas_cumprod)
show_points(steps, alphas_cumprod, "steps", "alphas_cumprod", "alphas_cumprod")
```

```python
test = np.sqrt((1. - alphas_cumprod) / alphas_cumprod)
# print(test)
tT= np.linspace(0., 1., T)
ms = tT
# ms = test ** 2 * tT
# ms = np.sqrt((1. - alphas_cumprod) / np.sqrt(alphas_cumprod))
plt.figure(figsize=(8,8))
plt.plot(steps, ms, marker='o', c='purple', markersize=0.1)
plt.plot(steps, test, marker='o', c='blue', markersize=0.1)
plt.show()
```

```python
sigma2 = 1. - alphas_cumprod - ms**2 * alphas_cumprod
#print(sigma2)
show_points(steps, sigma2, "steps", "sigma2", "sigma2")
```

```python
test = np.sqrt((1. - alphas_cumprod) / alphas_cumprod)
# print(test)
ms = np.arange(T)
ms = ms / (T - 1.)
plt.figure(figsize=(8,8))
plt.plot(steps, ms, marker='o', c='purple', markersize=0.1)
plt.plot(steps, test, marker='o', c='blue', markersize=0.1)
plt.show()
```

# BB_1000_linear_quard

```python
T = 1000
steps = np.arange(T)

ms = np.arange(T)
ms = ms / (T - 1)
ms[-1] = 1.00001
print(ms)
show_points(steps, ms, "steps", "ms", "ms")
```

```python
variance = np.arange(T)
variance = variance * (T-1. - variance) / (T-1.)
variance = variance * 0.25 * 4. / (T-1.) + 0.0001

print(variance)

show_points(steps, variance, "steps", "variance", "variance")
```

```python
sigma2_tminus_t = variance[1:T] - ((1. - ms[1:T]) / (1. - ms[0:T-1]))**2 * variance[0:T-1]
print(sigma2_tminus_t)
show_points(steps[1:T], sigma2_tminus_t, "steps", "sigma2_tminus_t", "sigma2_tminus_t")
```
